```
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```
#데이터 준비
file1 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'
lemonade = pd.read_csv(file1)
```

```
lemonade.head()
```

```
lemonade.shape
```

```
#x,y 쪼개기
```

```
feature = lemonade[['온도']]
```

```
target = lemonade[['판매량']]
```

```
print(feature.shape, target.shape)
```

```
#모델을 만듭니다.
X = tf.keras.layers.Input(shape=[1])
y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X,y)
model.compile(loss='mse')
```

```
#모델을 학습시킵니다.
model.fit(feature, target, epochs =1000)
```

```
model.fit(feature, target, epochs =10)
```

```
model.fit(feature, target, epochs =1000,verbose=0)
```

```
model.fit(feature, target, epochs =1000,verbose=0)
```

```
model.fit(feature, target, epochs =1000,verbose=0)
```

```
model.fit(feature, target, epochs =10)
```

```
#모델을 이용합니다.
print(model.predict(feature))
```
